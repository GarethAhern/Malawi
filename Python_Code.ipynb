```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster

import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\Gareth.Ahern\Desktop\data.csv')
data.head()
```

```python
m_marker = folium.Map(location=[50,-1], tiles='cartodbpositron', zoom_start=6)

# Add points to the map
for idx, row in data.iterrows():
    Marker([row['latitude'], row['longitude']]).add_to(m_marker)

# Display the map
#m_marker.save(r'C:\Users\Gareth.Ahern\Desktop\marker.html')
m_marker
```

```python
import math

m_cluster = folium.Map(location=[52,-0.5], tiles='cartodbpositron', zoom_start=5.5)

# Add points to the map
mc = MarkerCluster()
for idx, row in data.iterrows():
    if not math.isnan(row['longitude']) and not math.isnan(row['latitude']):
        mc.add_child(Marker([row['latitude'], row['longitude']]))
m_cluster.add_child(mc)

# Display the map
m_cluster.save(r'C:\Users\Gareth.Ahern\Desktop\cluster.html')
m_cluster
```

```python
# Create a base map
m_heat = folium.Map(location=[52,-0.5], tiles='cartodbpositron', zoom_start=5.5)

# Add a heatmap to the base map
HeatMap(data=data[['latitude', 'longitude']], radius=10).add_to(m_heat)

# Display the map
m_heat.save(r'C:\Users\Gareth.Ahern\Desktop\heat.html')
m_heat
```
